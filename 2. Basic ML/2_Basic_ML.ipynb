```python
players = pd.read_csv('/content/drive/MyDrive/YouTube Machine Learning/players.csv')

players.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
n, bins, patches = plt.hist(players['Height'], bins=5, color='blue', edgecolor='white', alpha=0.8)

for i in range(len(patches)):
    plt.text(patches[i].get_x() + patches[i].get_width() / 2, n[i],
             str(int(n[i])), ha='center', va='bottom', fontsize=10, color='white')

plt.title('Height Distribution of Wimbledon Finalists', fontsize=16, color='white')
plt.xlabel('Height (meters)', fontsize=12, color='white')
plt.ylabel('Frecuency', fontsize=12, color='white')
plt.grid(True, color='gray', linestyle='--', linewidth=0.5)
plt.xticks(ticks=bins, fontsize=10, color='white')

plt.gca().set_facecolor('black')
plt.gca().tick_params(colors='white')
plt.gcf().patch.set_facecolor('black')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(players['Birth Year'], players['Height'], color='blue', edgecolors='white', s=100, alpha=1)

plt.title('Height Evolution of Wimbledon Finalists', fontsize=16, color='white')
plt.xlabel('Birth Year', fontsize=12, color='white')
plt.ylabel('Height (meters)', fontsize=12, color='white')
plt.grid(True, color='gray', linestyle='--', linewidth=0.5)

plt.gca().set_facecolor('black')
plt.gca().tick_params(colors='white')
plt.gcf().patch.set_facecolor('black')

plt.show()
```
